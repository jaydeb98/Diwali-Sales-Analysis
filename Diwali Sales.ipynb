## 1. Number of Customers Based on Gender and Sales Made by Them 

```python
#Counting the number of customers based on gender.

gender_count = data.Gender.value_counts()

print(gender_count)
```

```python
#Extracting the sales based on gender.

sales_gen = data.groupby(['Gender'])['Amount'].sum()

print(sales_gen)
```

From the above numbers, we found that there are 7832 females and 3407 males who made purchases, and the respective sales made by them are 7,43,35,856 (approx) and 3,19,13,276. It showcases that, most buyers are females and they contributed the highest in overall sales. Now, we will display them using figures. 

```python
# Displaying the charts

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plotting the first chart: Gender Count
sns.barplot(x=gender_count.index, y=gender_count.values, ax=axs[0], palette=['pink', 'lightblue'])
axs[0].set_title('Gender Count')
axs[0].set_xlabel('Gender')
axs[0].set_ylabel('Count')

# Plotting the second chart: Sales by Gender
sns.barplot(x=sales_gen.index, y=sales_gen.values, ax=axs[1], palette=['pink', 'lightblue'])
axs[1].set_title('Sales by Gender')
axs[1].set_xlabel('Gender')
axs[1].set_ylabel('Total Sales Amount')

# Formatting the y-axis to show values in millions
axs[1].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x / 1e6))))

# Adding the legend
axs[1].legend(['Sales in Millions'], title='Note')

# Display the plots
plt.tight_layout()
plt.show()
```

## 2. Number of Customers Based on Age Group and Sales Made by Them

```python
#Counting the number of customers based on different age groups and gender.

age_group = data.groupby(['Age Group', 'Gender']).size().unstack()

print(age_group)
```

```python
#Extracting the sales based on different age groups and gender.

# Group by 'Age Group' and 'Gender' and sum the 'Amount'
sales_age = data.groupby(['Age Group', 'Gender'])['Amount'].sum().unstack().sort_index(ascending=False)

print(sales_age)
```

From the above numbers, we found that the age group of 26-35 made the highest purchases, followed by the age group 36-45 and 18-25. The least purchases made by the age group is 0-17. Females takes the lead in all sales made by each age-group. Now, we will display them using figures.

```python
# Displaying the charts

# Plotting the Age Group Counts by Gender
fig, ax = plt.subplots(figsize=(10, 6))
age_group.plot(kind='bar', stacked=False, ax=ax, color=['pink', 'lightblue'], width=0.7)
ax.set_title('Age Group Counts by Gender')
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()

# Plotting the Total Sales Amount by Age Group and Gender
fig, ax = plt.subplots(figsize=(10, 6))
sales_age.plot(kind='bar', stacked=False, ax=ax, color=['lightblue', 'pink'], width=0.7)
ax.set_title('Total Sales Amount by Age Group and Gender')
ax.set_xlabel('Age Group')
ax.set_ylabel('Total Sales Amount (in crores)')
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f"{x / 1e7:.2f} Cr"))

for container in ax.containers:
    labels = [f'{val / 1e7:.2f} Cr' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.show()
plt.tight_layout()
plt.show()
```

## 3. Number of Customers Based on Marital Status and Sales Made by Them

```python
# Add a new column 'Marital Status Description' based on the 'Marital Status' column.

data['Marital Status Description'] = data['Marital_Status'].apply(lambda x: 'Married' if x == 1 else 'Unmarried')
```

```python
data #To check the new column is added or not.
```

```python
#To count the number of married and unmarried customers.

marital_status_count = data['Marital Status Description'].value_counts()

print(marital_status_count)
```

```python
#Exctracting the sales based on marital status.

marital_status_sales = data.groupby(['Marital Status Description'])['Amount'].sum().sort_values(ascending=False)

print(marital_status_sales)
```

The above numbers tells that the number of unmarried customers are relatively higher than married ones, and unmarried customers contributed more to the overall sales in comparison to the married customers. Now we will display them in charts.

```python
# Displaying the charts

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plotting the first chart: Customers Count (Based on Marital Status)
sns.barplot(x=marital_status_count.index, y=marital_status_count.values, ax=axs[0], palette=['green', 'blue'])
axs[0].set_title('Customer Count (Based on Marital Status)')
axs[0].set_xlabel('Marital Status')
axs[0].set_ylabel('Count')

# Plotting the second chart: Sales Made by Customers (Based on Marital Status)
sns.barplot(x=marital_status_sales.index, y=marital_status_sales.values, ax=axs[1], palette=['green', 'blue'])
axs[1].set_title('Sales by Customers (Based on Marital Status)')
axs[1].set_xlabel('Marital Status')
axs[1].set_ylabel('Total Sales Amount (in crores)')
axs[1].get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: f"{x / 1e7:.2f} Cr"))
axs[1].legend(['Sales in Crores'], title='Note')


# Display the plots
plt.tight_layout()
plt.show()
```

## 4. Number of Customers by States and Sales Contribution of Each State 

```python
#Counting the number of customers based on states.

state_customers = data.State.value_counts()

print(state_customers)
```

```python
#Exctracting the sales based on each state.

state_sales = data.groupby(['State'])['Amount'].sum().sort_values(ascending = False)

print(state_sales)
```

From the above numbers, we can say that Uttar Pradesh is the state that has highest customers and also contributes the most on overall sales, followed by the states Maharastra, Karnataka, and Delhi. Now we will display them in figure.

```python
# Displaying the charts

# Plotting the number of customers based on states
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=state_customers.index, y=state_customers.values, ax=ax, palette='viridis')
ax.set_title('Number of Customers Based on States')
ax.set_xlabel('State')
ax.set_ylabel('Number of Customers')


for container in ax.containers:
    labels = [f'{int(val)}' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plotting the sales based on each state using Seaborn
fig, ax = plt.subplots(figsize=(10, 6))
state_sales_in_crores = state_sales / 1e7
sns.barplot(x=state_sales_in_crores.index, y=state_sales_in_crores.values, ax=ax, palette='viridis')
ax.set_title('Sales Based on Each State')
ax.set_xlabel('State')
ax.set_ylabel('Sales Amount (in crores)')

for container in ax.containers:
    labels = [f'{val:.2f} Cr' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## 5. Number of Customers Based on Occupation and Sales Made by Them 

```python
# Counting the number of customers based on occupation

occp_counts = data['Occupation'].value_counts().sort_values(ascending=False)

print(occp_counts)
```

```python
# Extracting the sales made by customers based on their occupation

sales_occp = data.groupby(['Occupation'])['Amount'].sum().sort_values(ascending = False)

print(sales_occp)
```

From the above numbers we can say that peoples who made the highest purchases are from IT Sector followed by Healthcare and Aviation industry. Now, we will demonstrate this by using figures. 

```python
# Displaying the charts

# Plotting the number of customers based on their occupation
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=occp_counts.index, y=occp_counts.values, ax=ax, palette='coolwarm')
ax.set_title('Number of Customers Based on Their Occupation')
ax.set_xlabel('Occupation')
ax.set_ylabel('Number of Customers')

for container in ax.containers:
    labels = [f'{int(val)}' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plotting the sales based on occupation
fig, ax = plt.subplots(figsize=(10, 6))
sales_occp_in_millions = sales_occp / 1e7
sns.barplot(x=sales_occp_in_millions.index, y=sales_occp_in_millions.values, ax=ax, palette='coolwarm')
ax.set_title('Sales Based on Occupation')
ax.set_xlabel('Occupation')
ax.set_ylabel('Sales Amount (in crores)')

for container in ax.containers:
    labels = [f'{val:.2f} Cr' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## 6. Number of Products Sold Based on Category and Sales Contribution of Each Product 

```python
# Counting the total number of products sold based on each categories

product_sold = data.Product_Category.value_counts()

print(product_sold)
```

```python
# Sales contribution of each product categories

product_sales = data.groupby(['Product_Category'])['Amount'].sum().sort_values(ascending=False)

print(product_sales)
```

From the above numbers we saw some interesting pattern, Clothing & Apparel is the top-selling product category in terms of quantity. However, the Food category leads in overall sales revenue, generating more than twice the amount of sales compared to the Clothing & Apparel category. Now, let's demonstrate them in figures. 

```python
# Displaying the charts

# Plotting the number of total number of products sold based on each category
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=product_sold.index, y=product_sold.values, ax=ax, palette='cividis')
ax.set_title('Number of Products Sold Based on Each Category')
ax.set_xlabel('Product Category')
ax.set_ylabel('Number of Product Sold')

for container in ax.containers:
    labels = [f'{int(val)}' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plotting the sales based on each category
fig, ax = plt.subplots(figsize=(10, 6))
product_sales_in_millions = product_sales / 1e7
sns.barplot(x=product_sales_in_millions.index, y=product_sales_in_millions.values, ax=ax, palette='cividis')
ax.set_title('Sales Based on Product Category')
ax.set_xlabel('Product Category')
ax.set_ylabel('Sales Amount (in crores)')

for container in ax.containers:
    labels = [f'{val:.2f} Cr' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
